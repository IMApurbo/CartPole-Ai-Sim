```python
import gymnasium as gym  # Gymnasium replaces the older Gym
from stable_baselines3 import PPO  # Replace PPO with another algorithm if needed
```

```python
env = gym.make("CartPole-v1")
```

```python
obs, info = env.reset()  # Unpack observation and info
```

```python
model = PPO("MlpPolicy", env, verbose=1)
```

```python
# Train the model
model.learn(total_timesteps=100000)
```

```python
model.save('Cartpole-Ai-Model')
```

```python
model = PPO.load('Cartpole-Ai-Model')
```

```python
env = gym.make("CartPole-v1", render_mode="human")
obs, info = env.reset()  # Reset environment
for _ in range(1000):  # Evaluate for 1000 steps
    action, _states = model.predict(obs, deterministic=True)  # Predict action
    obs, reward, terminated, truncated, info = env.step(action)  # Step through environment
    env.render()  # Render the environment
    if terminated or truncated:  # Handle termination or truncation
        obs, info = env.reset()
```
